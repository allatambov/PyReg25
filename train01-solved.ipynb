### Задача 2

a. Сохраните в массив `kr1` баллы за контрольную работу от 30 и выше.

c. Сохраните в массив `kr3` баллы за контрольную работу от 10 до 25 включительно.

d. Вычислите число студентов, которые не писали работу, т.е. тех, чей балл равен 0.

e. Создайте массив `kr_bin` из 1 и 0, где 1 соответствует «зачетному» баллу (не ниже 17).

```python
kr = np.array([29.5, 21.5, 43.5, 20.5, 40.5, 24.5, 17.25, 8.0, 16.75, 
               10.5, 19.0, 39.0, 34.0, 15.25, 28.5, 39.0, 33.0, 40.5, 
               38.0, 19.0, 1.0, 29.0, 42.0, 36.0, 38.5, 36.0, 35.5, 
               48.0, 8.5, 10.5, 11.0, 0.0, 36.5, 23.0, 45.0, 33.0, 
               0.0, 6.0, 23.5, 19.5, 12.0, 28.5, 44.5, 28.5, 32.0, 
               39.0, 0.0, 40.0, 27.0, 31.0, 23.5, 37.0, 21.75, 24.0, 
               18.25, 28.5, 27.0, 9.5, 0.0, 4.0, 4.0, 13.5, 14.5, 27.0, 
               29.0, 3.0, 36.5, 32.5, 13.5, 0.0, 12.0, 4.5, 36.0, 38.0, 
               25.0, 29.5, 39.0, 26.0, 11.0, 48.5, 21.5, 11.5, 17.0, 
               34.0, 15.0, 30.5, 36.0, 28.0, 11.0, 25.0, 44.0, 42.5, 
               21.5, 29.5, 8.0])
```

```python
# a
kr1 = kr[kr >= 30]
print(kr1)
```

```python
# b
kr2 = kr[kr < 20]
print(kr2)
```

```python
# c
kr3 = kr[(kr >= 10) & (kr <= 25)] 
print(kr3)
```

```python
# d

zeroes = sum(kr == 0)
print(zeroes)
print(len(kr[kr == 0]))
```

```python
# e: вариант с astype()

kr_bin = (kr >= 17).astype(int)
print(kr_bin)
```

```python
# e: вариант с where()

kr_bin = np.where(kr >= 17, 1, 0)
print(kr_bin)
```

### Задача 3

Используя массив `kr` из предыдущей задачи и строки кода ниже:

a. сохраните в массив `sample01` значения, соответствующие студентам, чей балл не ниже среднего;

b. сохраните в массив `sample02` значения, соответствующие студентам, чей балл выше нижнего квартиля, но ниже медианы;

c. сохраните в массив `sample03` значения, соответствующие студентам, чей балл выше медианы, но ниже верхнего квартиля.

```python
# медиана, нижний и верхний квартиль

med = np.median(kr)
q1 = np.quantile(kr, 0.25)
q3 = np.quantile(kr, 0.75)
```

```python
# варианты для среднего

mean = sum(kr) / len(kr)
print(mean, np.mean(kr), kr.mean())
```

```python
# a
sample01 = kr[kr >= mean]
print(sample01)
```

```python
# b
sample02 = kr[(kr > q1) & (kr < med)]
print(sample02)
```

```python
# c
sample03 = kr[(kr > med) & (kr < q3)]
print(sample03)
```

### Задача 4

На основе массива `kr` из задачи 3 вычислите оценки за контрольную работу в 10-балльной шкале, считая, что оценка получалась нормированием на максимально набранный балл (посчитан ниже).

```python
max_score = np.max(kr)
```

```python
score = kr / max_score * 10
print(score)
```

```python
# дополнение: округляем оценки и переводим в integer

np.round(score).astype(int)
```
