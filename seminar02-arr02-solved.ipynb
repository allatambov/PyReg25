### Задача 2

В массиве `group` зафиксировано, к какой группе относится участник эксперимента, посвященного эффективности медицинского препарата. Группы A и B являются тестовыми – участники принимают препарат в разных дозировках, а группа C является контрольной – участники принимают плацебо. 

a. Выберите из массива `group` представителей групп A и B и сохраните их в массив `test`, остальных сохраните в массив `control`. 

b. На основе `group` создайте массив `status` из 0 и 1, где 1 соответствует тестовой группе, а 0 – контрольной.

```python
group = np.array(["A", "A", "A", "B", "C", "A", "B", "B", "B", "B", 
                  "C", "A", "A", "C", "C", "C", "A", "B", "B", "A", 
                  "A", "A", "B", "B", "B", "C", "A", "A", "B", "C",
                  "B", "C", "A", "B", "A", "C", "B", "B", "C", "B", 
                  "B", "C", "A", "C", "C", "A", "A", "C", "B", "A"])
```

```python
# a: вариант 1
# группа A ИЛИ группа B -> test,
# группа C -> control

test = group[(group == "A") | (group == "B")]
control = group[group == "C"]

print(test)
print(control)
```

```python
# a: вариант 2 – если хочется использовать отрицание
# так как групп всего 3, в test помещаем всё, что не C
# а для control строим отрицание через ~

test = group[group != "C"]
control = group[~((group == "A") | (group == "B"))]

print(test)
print(control)
```

```python
# b: вариант 1
# astype(int) превращает набор из True/False в набор целых чисел,
# переводим тип bool в int

status = ((group == "A") | (group == "B")).astype(int)
print(status)
```

```python
# b: вариант 2
# функция np.where(): на первом месте условие,
# на втором – что ставить, если условие выполняется,
# на третьем – что ставить, если условие не выполняется

status = np.where((group == "A") | (group == "B"), 1, 0)
print(status)
```
